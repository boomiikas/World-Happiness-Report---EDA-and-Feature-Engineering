```python
df['Ladder score'].isna().sum()
```

```python
df.loc[:, 'Ladder score'] = pd.to_numeric(df['Ladder score'], errors='coerce')
if df['Ladder score'].notna().any():
    median_val = df['Ladder score'].median(skipna=True)
else:
    median_val = 0 
df.loc[:, 'Ladder score'] = df['Ladder score'].fillna(median_val)
```

```python
df.loc[:, 'Log GDP per capita'] = pd.to_numeric(df['Log GDP per capita'], errors='coerce')
if df['Log GDP per capita'].notna().any():
    median_val = df['Log GDP per capita'].median(skipna=True)
else:
    median_val = 0
df.loc[:, 'Log GDP per capita'] = df['Log GDP per capita'].fillna(median_val)
```

```python
ladder_score = np.array(df['Ladder score'])
logged_gdp_per_capita = np.array(df['Log GDP per capita'])
```

```python
vec_add=ladder_score+logged_gdp_per_capita
vec_sub=ladder_score-logged_gdp_per_capita
print("Add:",vec_add)
print("\n")
print("Sub:",vec_sub)
```

```python
print(round(np.dot(ladder_score,logged_gdp_per_capita),2))
```

```python
import sympy as sp
```

```python
GDP, H = sp.symbols('GDP H')
Wellbeing_Index = (GDP * H) + 0.5 * (H - 5)**2
derivative = sp.diff(Wellbeing_Index, H)
print("Derivative:", derivative)
```
